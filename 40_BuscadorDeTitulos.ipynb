```python
from PyPDF2 import PdfFileReader
```

```python
temp = open('Cancionero.pdf', 'rb')
PDF_read = PdfFileReader(temp)
number_pages = PDF_read.getNumPages()
print('El numero de paginas del documento es: {}'.format(number_pages))
print()

indice = ''
for i in range(11):
    page = PDF_read.getPage(i)
    page_extract = page.extractText()
    indice = indice + page_extract
    print(page_extract) #Impresion en pantalla
    
    
```

```python
indice
```

```python
import re
corta = re.split(r'[0123456789.]\s*',indice)
longitud = len(corta)
print('longitud de corta {}'.format(longitud))

print(corta)
```

```python
seleccion = []
for iterador in range(0,longitud):
    if corta[iterador]!='':
        seleccion.append(corta[iterador])

print(seleccion)
```

```python
long = len(seleccion)
print(long)
```

```python
seleccion2 = []
for iterador in range(0,long):
    if seleccion[iterador]!='………………………………………………………………………………………………………………  ':
        seleccion2.append(seleccion[iterador])

print(seleccion2)
```
