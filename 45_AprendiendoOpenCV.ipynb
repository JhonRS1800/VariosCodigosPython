### Leyendo una imagen de color en escala grises

```python
import numpy as np
import cv2

# Cargando una imagen de color en escala grises
img = cv2.imread("C:/Users/jon_r/Pictures/FOTOS/Mis Fotos/image10.jpg",0)

# Muestra una imagen
cv2.imshow('imagen', img)

# Tiempo de espera hasta que se presione una tecla
cv2.waitKey(0)

# Elimina las ventanas emergentes
cv2.destroyAllWindows()
```

### Guardando una imagen

```python
import numpy as np
import cv2

# Cargando una imagen de color en escala grises
img = cv2.imread("C:/Users/jon_r/Pictures/FOTOS/Mis Fotos/image10.jpg",0)

# Muestra una imagen
cv2.imshow('imagen.jpg', img)

# Guardando la imagen
cv2.imwrite('amigos_grises.jpg', img)

# Tiempo de espera hasta que se presione una tecla
cv2.waitKey(0)

# Elimina las ventanas emergentes
cv2.destroyAllWindows()
```

### Usando Matplotlib

Ocultando los valores de marca de los ejes X & Y

```python
import numpy as np
import cv2
from matplotlib import pyplot as plt

img = cv2.imread("C:/Users/jon_r/Pictures/FOTOS/Mis Fotos/image10.jpg", 0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([]) # Para ocultar los valores de marca en los ejes x & y
plt.show()
```
